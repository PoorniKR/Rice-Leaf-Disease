```python
import h5py

with h5py.File("inception_features.h5", "r") as f:
    X_train_inception = f["X_train"][:]
    X_val_inception = f["X_val"][:]
    X_test_inception = f["X_test"][:]

print("Extracted IRV2 features loaded from HDF5!")
```

```python
with h5py.File("vit_features.h5", "r") as f:
    X_train_vit = f["X_train"][:]
    X_val_vit = f["X_val"][:]
    X_test_vit = f["X_test"][:]

print("Extracted ViT features loaded from HDF5!")
```

```python
import numpy as np

w_inception = 0.7
w_vit = 0.3

X_train_fused = np.concatenate([w_inception * X_train_inception, w_vit * X_train_vit], axis=1)
X_val_fused = np.concatenate([w_inception * X_val_inception, w_vit * X_val_vit], axis=1)
X_test_fused = np.concatenate([w_inception * X_test_inception, w_vit * X_test_vit], axis=1)
```

```python
import h5py

with h5py.File("fused_features.h5", "w") as f:
    f.create_dataset("X_train", data=X_train_fused)
    f.create_dataset("X_val", data=X_val_fused)
    f.create_dataset("X_test", data=X_test_fused)

print("Features saved in HDF5 format!")
```

```python
with h5py.File("fused_features.h5", "r") as f:
    X_train_fused = f["X_train"][:]
    X_val_fused = f["X_val"][:]
    X_test_fused = f["X_test"][:]

print("Extracted fused features loaded from HDF5!")
```

```python
print("Train Features Shape:", X_train_fused.shape)
print("Validation Features Shape:", X_val_fused.shape)
print("Test Features Shape:", X_test_fused.shape)

print("Feature Data Type:", X_train_fused.dtype)
```
